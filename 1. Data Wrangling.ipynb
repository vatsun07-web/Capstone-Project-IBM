<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
df['ConvertedCompYearly'].describe()
```

```python
df['ConvertedCompYearly'].isnull().sum()
```

```python
comp_mode = df['ConvertedCompYearly'].median() 
df['ConvertedCompYearly'].fillna(comp_mode, inplace =True)
if df['ConvertedCompYearly'].isnull().sum() == 0 :
    print('The ConvertedCompYearly was imputed missing value')
```

```python
df['ConvertedCompYearly'].isnull().sum()
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
Remote_freq = df['RemoteWork'].value_counts().idxmax() 
Remote_freq
```

```python
df['RemoteWork'].isnull().sum()
```

```python
df['RemoteWork'].fillna(Remote_freq, inplace = True) 
if df['RemoteWork'].isnull().sum() == 0:
    print('The RemoteWork was imputed missing value')
```

### 6. Feature Scaling and Transformation

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
df['YearsCodePro'].describe()
```

```python
df['YearsCodePro'].info()
```

```python
df[df['YearsCodePro'] == 'Less than 1 year']
```

```python
df['YearsCodePro'].replace('Less than 1 year', 1, inplace =True)
```

```python
df['YearsCodePro'].fillna(0, inplace =True) 
```

```python
df['YearsCodePro'].replace('More than 50 years', 51 , inplace =True)
```

```python
df['YearsCodePro'] = df['YearsCodePro'].astype(int)
```

```python
bins = np.linspace(min(df['YearsCodePro']), max(df['YearsCodePro']), 4) 
group = ['Junior', 'Mid-level', 'Senior/Lead'] 
df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], bins, labels= group, include_lowest = True)
```
